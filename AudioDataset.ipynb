```python
class SpleeterDataset(Dataset):

    def __init__(self, dataset_type):
        mus = musdb.DB(subsets=dataset_type, root='./dataset')
        self.tracks=[]
        self.vocals=[]
        for track in mus :
            track.chunk_duration = 5.0
            track.chunk_start = random.uniform(0, track.duration - track.chunk_duration)
            x = track.audio.T
            y = track.targets['vocals'].audio.T
            self.tracks.append(x)
            self.vocals.append(y)
    
    def __len__(self):
        return len(self.vocals)
        

    def __getitem__(self, idx):
        return self.tracks[idx], self.vocals[idx]
```

```python
train = SpleeterDataset("train")
test = SpleeterDataset("test")
```

```python
train_dataloader = DataLoader(train, batch_size=4, shuffle=True)
test_dataloader = DataLoader(test, batch_size=4, shuffle=True)
```
